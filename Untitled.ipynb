```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# 1. Membuat SparkSession
spark = SparkSession.builder \
    .appName("Contoh PySpark Sederhana") \
    .getOrCreate()

# 2. Membuat DataFrame dari data list
data = [
    ("Andi", 25),
    ("Budi", 30),
    ("Citra", 22),
    ("Dina", 28)
]

columns = ["Nama", "Umur"]

df = spark.createDataFrame(data, columns)

# 3. Menampilkan isi DataFrame
print("Data Awal:")
df.show()

# 4. Filter data (misalnya umur > 25)
df_filtered = df.filter(col("Umur") > 25)

print("Data setelah filter umur > 25:")
df_filtered.show()
```

```python
# 5. Menutup SparkSession
spark.stop()
```
